```python
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.datasets import make_regression
from sklearn.linear_model import LinearRegression
```

```python
#Regression
x,y = make_regression(n_samples=200, n_features=1, noise=30)
#make_regression cria uma amostra de dados aleatorios
plt.scatter(x,y) # plotar amostra
```

```python
modelo = LinearRegression()
modelo.fit(x,y)
```

```python
print('Coeficiente Angular: {0}\nCoeficiente Linear:{1}'.format(modelo.coef_,modelo.intercept_))
```

```python
a_coeff = modelo.coef_
l_coeff = modelo.intercept_
# y = a + b*x
y = l_coeff + a_coeff * x
```

```python
plt.scatter(x,y)
plt.plot(x, l_coeff + a_coeff*x , color='red')
plt.scatter(0, l_coeff + a_coeff*0, color='green', s=300)
plt.show()
```
